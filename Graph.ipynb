```python
%matplotlib inline
import matplotlib.pyplot as pt

x=np.arange(1,10)
y=x**2
pt.plot(x,y)
#pt.show()
```

```python
pt.scatter(x,y)
```

```python
pt.plot(x,y,color='g')
pt.xlabel('sales')
pt.ylabel('count')
```

```python
pt.subplot(2,2,1)
pt.plot(x,y,color='g',linewidth=20,alpha=.7)
pt.xlabel('sales')
pt.subplot(2,2,2)
pt.plot(x,y,color='k')
pt.ylabel('count')
pt.subplot(2,2,3)
pt.plot(x,y,color='g')
pt.subplot(2,2,4)
pt.plot(y,x,color='g')
```
