# Exponential Smoothing Model

```python
train.shape[0]
```

```python
a=0.6
x=[]
for i in range(0,train.shape[0]):
    x.append(a*(1-a)**i)
    
list(reversed(x))
train['Exp_Smoothing_Coef']=list(reversed(x))
```

```python
train.head()
```

```python
train['Sales_Exp']=train['Sales']*train['Exp_Smoothing_Coef']
```

```python
train['Sales_Exp'].mean()
```

```python
test['Sales1']=train['Sales_Exp'].mean()
test['Error']=test['Sales']-test['Sales1']
```

```python
test.plot()
```
