```python
import cv2

# Load our input image
image = cv2.imread("C:\\Users\\29265\\Downloads\\bike images _ Google Search\\abc.jpg")
cv2.imshow('Original', image)
cv2.waitKey()
B,G,R=cv2.split(image)
cv2.imshow('Blue', B)
cv2.waitKey()
cv2.imshow('Green', G)
cv2.waitKey()
cv2.imshow('red', R)
cv2.waitKey()
cv2.destroyAllWindows()
mergerd=cv2.merge([B,G,R])
cv2.imshow('mergerd', mergerd)
cv2.waitKey()

mergerd1=cv2.merge([B+10,G,R])
cv2.imshow('changed', mergerd1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#to Set yellow color for even pixel
import cv2
row, column, channels = image.shape
 
# change all pixels to random value
for i in range(row):
    for j in range(column):
        a=image[i, j]
        if (a[0] % 2) == 0 and (a[1] % 2) == 0 and (a[2] % 2) == 0:
            B_random = 0
            G_random = 255
            R_random = 255
            image[i, j] = [B_random, G_random, R_random]
 
# show image
#cv2.imwrite('output.png', image)
cv2.imshow("Image", image)
cv2.waitKey()
cv2.destroyAllWindows()
```
