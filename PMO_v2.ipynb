```python
def findAllIndex(df,key):
    df=df.groupby(key)
    return list(df.sum().index)
```

```python
def formatDate(str1):
    str1=str1.split(" ")
    str1=str1[0]
    str1=str1.split("-")
    Srt1=str1[1]+"/"+str1[2]+"/"+str1[0]
    return Srt1
```
