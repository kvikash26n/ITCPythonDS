```python
#for i in dfread_inputS1.columns:
list(dfread_inputS1.columns)  

data_col_inputS1=list(dfread_inputS1.columns)
data_col_outputS1=list(dfread_outputS1.columns)

data_col_inputS2=list(dfread_inputS2.columns)
data_col_outputS2=list(dfread_outputS2.columns)

def Unmatched(x,y):
    return list(set(x)-set(y))

    
```

```python
#columns not match value from here
NotMatchedColumns1=Unmatched(data_col_inputS1,data_col_outputS1)
NotMatchedColumns2=Unmatched(data_col_inputS2,data_col_outputS2)
#print(NotMatchedColumns1)
#print(NotMatchedColumns2)
```

```python
def GetColum(NotMatchedColumns1):
    
    coln=pd.DataFrame(NotMatchedColumns1)
    coln.rename(columns={0:'ColumnName'},inplace=True)
    return coln
```

```python
#get Columns not match val for s1,s2
ns1=GetColum(NotMatchedColumns1)
ns2=GetColum(NotMatchedColumns2)
ns2
```

```python
#data from each column
#colvalue_not_matched=[]
def getColumnWiseNotMatchVal(dfread_inputS1,dfread_outputS1):
    colvalue_not_matched=[]
    for i in dfread_inputS1.columns:
        for j in dfread_outputS1.columns:
            if i==j:
                a=list(dfread_inputS1[i])
                b=list(dfread_outputS1[i])
                notmatchval=Unmatched(a,b)
                if len(notmatchval)!=0:


                    dict={
                    'columns':i,
                        'val':notmatchval
                    }
                    colvalue_not_matched.append(dict)
    colvaluenotmatch=pd.DataFrame(colvalue_not_matched)
    return colvaluenotmatch
#print(colvalue_not_matched)
```

```python
s1=getColumnWiseNotMatchVal(dfread_inputS1,dfread_outputS1)
s2=getColumnWiseNotMatchVal(dfread_inputS2,dfread_outputS2)
```
