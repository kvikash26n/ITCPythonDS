```python
def findDay(dateinput):
    import datetime
    a=dateinput #MM/DD/YYYY
    a=a.split("/")
    b=a[1]+"/"+a[0]+"/"+a[2]

    dt =b #'01/04/2019' # dd/mm/YYYY
    day, month, year = (int(x) for x in dt.split('/'))    
    ans = datetime.date(year, month, day)
    #print (ans.strftime("%A"))
    day=ans.strftime("%A")
    return day
```
