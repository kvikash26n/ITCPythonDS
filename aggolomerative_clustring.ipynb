```python
import pandas as pd
df=pd.read_csv("C:\\Users\\29265\\Desktop\\train_ross.csv")
```

```python
def create_tab(df,c_1,c_2,c_3):
    c_1=c_1
    c_2=c_2
    c_3=c_3
    #df_new=df.loc[0:,['Store','Sales','Customers']]
    df_new=df.loc[0:,[c_1,c_2,c_3]]
    return df_new
```

```python
df_new=create_tab(df,'Store','Sales','Customers')
```

```python
df_new=df_new.sample(6)
```

```python
df_new
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
dendrogram = sch.dendrogram(sch.linkage(df_new, method = 'single'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.show()
```
