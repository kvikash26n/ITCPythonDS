```python
import pandas as pd
import numpy as np
x=[] 
for j in range(0,6):
    test=pd.DataFrame(np.random.rand(100,4),columns=['sales','gdp','tansx','profit'])
    control=pd.DataFrame(np.random.rand(100,4),columns=['sales','gdp','tansx','profit'])
    a=[]
    for i in test.columns:
        a.append(test[i].sum())
    b=[]
    for i in control.columns:
        b.append(control[i].sum())
    a=np.array(a)
    b=np.array(b)
    xz=((a-b)/b)*100
    x.append(xz)
x
```

```python
lift_table=pd.DataFrame(x,columns=['sales','gdp','tansx','profit'])#,index=['m1','m2','m3','m4','m5','m6'])
```

```python
lift_table
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

for i in lift_table.columns:
    plt.xlabel('month')
    plt.ylabel('sales data')
    plt.plot(lift_table[i])
    
    plt.show()
```
