```python
#a=df.iloc['Store',]
def create_tab(df,c_1,c_2,c_3):
    c_1=c_1
    c_2=c_2
    c_3=c_3
    #df_new=df.loc[0:,['Store','Sales','Customers']]
    df_new=df.loc[0:,[c_1,c_2,c_3]]
    return df_new
```

```python
df_new=create_tab(df,'Store','Sales','Customers')
```

```python
df_new
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
kmeans=KMeans(n_clusters=3)
kmeans.fit(df_new)
```

```python
centroids=kmeans.cluster_centers_
```

```python
labels=kmeans.labels_
```
